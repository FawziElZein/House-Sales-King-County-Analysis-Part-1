### Convert from Series to Data frame:

```python
# resetting display option since we no more need to display a hug number of rows.
pd.reset_option("display")
```

```python
df_built_vs_price['price'].dtypes # This is a serie data type showing all prices with labels on the far left
```

```python
price_occurence = df_built_vs_price['price'].value_counts() # This is also a series data type showing the prices and the number of occurrences
price_occurence
```

```python
#Convert series into a data frame, by default if no parameter was passed to the function, the column name will be set as 
#the name of the series (in this case it's count)
price_occurence.to_frame()
```

```python
# In case we would like to change the name then we pass name="freq_price" as a parameter to the function below 
price_occurence.to_frame(name="freq_price")
```
